```python
p0  = dfe["total_time"].mean()
p1 =  dfe["total_time"].std()

p2  = dfe["avg_time"].mean()
p3 =  dfe["avg_time"].std()

p4  = dfe["std_time"].mean()
p5 =  dfe["std_time"].std()
```

```python
def simple_check (val, params, n=1.0):
    #unpack
    p1, p2, p3, p4, p5, p6 = params

    value = False

    v_total = np.sum(val)
    v_avg = np.mean(val)
    v_std = np.std(val)

    if v_total >= (p1 - n*p2) and v_total <= (p1 + n*p2):
        value = True

    if v_avg >= (p3 - n*p4) and v_avg <= (p3 + n*p4):
        value = value and True

    if v_std >= (p5 - n*p6) and v_std <= (p5 + n*p6):
        value = value and True


    return value
```

```python
df.loc[0]
```

```python
simple_check(df.loc[3], (p0,p1,p2,p3,p4,p5), n=1.0)
```

```python
for i in range(20):
    print("Case: ",df.loc[i], ", pass: ", simple_check(df.loc[i], (p0,p1,p2,p3,p4,p5), n=1.0))
```

```python
import json

def generate_json(dataframe):
    dfe = dataframe.copy()
    dfe["total_time"] = dataframe.sum(axis=1)
    dfe["avg_time"] = dataframe.sum(axis=1)/len(p_str)
    dfe["std_time"] = dataframe.std(axis=1)
    
    params = {
        "type": "stadistical_basic",
        "version": "0.1",
                "p": [
                    (dfe["total_time"].mean()),
                    (dfe["total_time"].std()),
                    (dfe["avg_time"].mean()),
                    (dfe["avg_time"].std()),
                    (dfe["std_time"].mean()),
                    (dfe["std_time"].std())
                ],
                "n": 1.0,
                "v": "[sum(i), sum(i)/len(i), std(i)];",
                "criteria": "(v[0]>=(p[0]-n*p[1]) && v[0]<= (p[0] + n*p[1])) && (v[1]>=(p[2]-n*p[3]) && v[1]<= (p[2] + n*p[3])) && (v[2]>=(p[4]-n*p[5]) && v[3]<= (p[4] + n*p[5]))"
    }
    return json.dumps(params)
```
